# Question 2: Price Distribution



```python
plt.figure(figsize=(15,12))
plt.hist(df['price'], bins=50)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

# Question 3: Price Distribution by brand

```python
df_grouped = df.groupby('brand')
plt.figure(figsize=(16,12))
for brand, group in df_grouped:
    plt.hist(group['price'], bins=50, alpha=0.5, label=brand)

plt.xlabel('Price')
plt.ylabel('Frequency')
plt.legend(loc='upper right')
plt.show()
```

**plot of each brand for better view**

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(nrows=len(df['brand'].unique()), figsize=(12, 30))

for i, brand in enumerate(df['brand'].unique()):
    axs[i].hist(df[df['brand']==brand]['price'], bins=50, alpha=0.5)
    axs[i].set_title(brand)

plt.tight_layout()
plt.show()
```

# Question 4: Correlation betwwen price and rating

**Correlation plot between numeric valirable**

As we can see on the correlation matrix, correlation score between price and rating is 0.11, which is low


Which mean there is a weak linear relationship between the two columns, or the high price does not really associate with high rating

```python
import seaborn as sns


# Select the columns to plot
cols = ['rating', 'num_ratings', 'price', 'shipping_fee']

# Create a pair plot
sns.pairplot(df[cols])

# Adjust the figure size
fig = plt.gcf()
fig.set_size_inches(15, 15)

# Display the plot
plt.show()
```

```python
corr = df[cols].corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(10,8))
sns.heatmap(corr, cmap='coolwarm', annot=True, fmt='.2f', square=True)
plt.title('Correlation Matrix')
plt.show()
```
